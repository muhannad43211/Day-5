#### Q4: Create a function that displays books information from your library and test it.

```python
# write your code here ^_^
def display_books(library):

    if not library:
        print("The library is empty.")
        return

    header = ["Book ID", "Title", "Author", "Year", "Genres"]
    print("{:<10} {:<50} {:<30} {:<10} {:<30}".format(*header))
    print("=" * 120)

    for book in library:
        genres_str = ", ".join(book["Genres"])
        print("{:<10} {:<50} {:<30} {:<10} {:<30}".format(
            book["Book ID"], book["Title"], book["Author"], book["Year"], genres_str
        ))

display_books(library)
```

#### Q5: Create functions to do the following: calculates and display the total number of books, the average publication year, and the most common genre. and test it.

```python
# write your code here ^_^
def calculate_total_books(library):

    return len(library)

def calculate_average_publication_year(library):

    if not library:
        return 0

    total_years = sum(book["Year"] for book in library)
    return total_years / len(library)

def find_most_common_genre(library):

    if not library:
        return None

    genre_counts = {}
    for book in library:
        for genre in book["Genres"]:
            genre_counts[genre] = genre_counts.get(genre, 0) + 1

    most_common_genre = max(genre_counts, key=genre_counts.get)
    return most_common_genre


total_books = calculate_total_books(library)
average_year = calculate_average_publication_year(library)
most_common = find_most_common_genre(library)

print("Total number of books:", total_books)
print("Average publication year:", average_year)
print("Most common genre:", most_common)
```

#### Q6: Create 2 functions to search for books by the title and the author from your library and test it.

```python
# write your code here ^_^
def search_by_title(library, title):

    results = []
    for book in library:
        if title.lower() in book["Title"].lower():
            results.append(book)
    return results

def search_by_author(library, author):

    results = []
    for book in library:
        if author.lower() in book["Author"].lower():
            results.append(book)
    return results

title_search_results = search_by_title(library, "Harry Potter")
author_search_results = search_by_author(library, "J.K. Rowling")

print("Search results by title:")
display_books(title_search_results)

print("\nSearch results by author:")
display_books(author_search_results)
```

#### Q7: Create 2 functions to sort the books by the title and the year from your library and test it.

```python
# write your code here ^_^
def sort_by_title(library):
  
    library.sort(key=lambda book: book["Title"].lower())

def sort_by_year(library):

    library.sort(key=lambda book: book["Year"])

sort_by_title(library)
print("Sorted by title:")
display_books(library)

sort_by_year(library)
print("\nSorted by year:")
display_books(library)
```

#### Q8: Create a function to bulk update genres of books using list comprehension. and test it.

```python
# write your code here ^_^
def update_genres_bulk(library, old_genre, new_genre):

    for book in library:
        book["Genres"] = [new_genre if genre == old_genre else genre for genre in book["Genres"]]

update_genres_bulk(library, "Classics", "Literature")
print("Genres updated:")
display_books(library)
```

#### Q9: Implement a function to generate a report summarizing the library's statistics, including the total number of books, the number of books by each author, the number of books in each genre, and the oldest and newest books. and test it.

```python
# write your code here ^_^
def generate_library_report(library):

    total_books = len(library)
    author_counts = {}
    genre_counts = {}
    oldest_book = None
    newest_book = None

    for book in library:
        author_counts[book["Author"]] = author_counts.get(book["Author"], 0) + 1
        for genre in book["Genres"]:
            genre_counts[genre] = genre_counts.get(genre, 0) + 1

        if not oldest_book or book["Year"] < oldest_book["Year"]:
            oldest_book = book
        if not newest_book or book["Year"] > newest_book["Year"]:
            newest_book = book

    print("Library Report")
    print("-----------------")
    print("Total number of books:", total_books)
    print("\nNumber of books by author:")
    for author, count in author_counts.items():
        print(f"{author}: {count}")
    print("\nNumber of books by genre:")
    for genre, count in genre_counts.items():
        print(f"{genre}: {count}")
    print("\nOldest book:")
    display_books([oldest_book])
    print("\nNewest book:")
    display_books([newest_book])

generate_library_report(library)
```
